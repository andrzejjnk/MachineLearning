```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
#Zadanie 1
iris
```

```python
#Zadanie 2
rows = iris.shape[0]
columns = iris.shape[1]
print(f"Number of rows {rows}, columns {columns}")
```

```python
#Zadanie 3
iris.describe()
```

```python
#Zadanie 4
iris.groupby(by=['sepal length (cm)']).mean()
```

```python
#Zadanie 5
iris.head(5)
```

```python
#Zadanie 6
count_NaN = iris.isnull().sum()
print(f"Missing values in each column of dataframe:\n{count_NaN}")
iris.dropna(inplace=True)
iris
```

```python
#Zadanie 7
iris.sort_values(by=['sepal width (cm)'], ascending=True, inplace=True)
iris
```

```python
#Zadanie 8
min_petal_length = iris['petal length (cm)'].min()
min_petal_length_idx = iris['petal length (cm)'].idxmin()
max_petal_length = iris['petal length (cm)'].max()
max_petal_length_idx = iris['petal length (cm)'].idxmax()
print(f"Minimum value in column named petal length {min_petal_length}, index {min_petal_length_idx}.")
print(f"Maximum value in column named petal length {max_petal_length}, index {max_petal_length_idx}.")
```

```python
#Zadanie 9
iris.std()
iris
```

```python
#Zadanie 10
average_length = iris['sepal length (cm)'].mean()
print(f"Average length = {average_length}")
```

```python
iris_result = iris.loc[iris['sepal length (cm)'] > average_length]
iris_result
```

```python
#Zadanie 11
iris_columns = list(iris.columns[0:-1])
legend_labels = list(set(iris['target']))
fig, axs = plt.subplots(2,2, figsize=(30,20))
x = iris.pivot(columns='target', values=iris_columns[0])
axs[0,0].hist(x, bins=30)
axs[0,0].set_title("sepal length (cm)")
axs[0,0].set_ylabel("sepal length (cm)")
axs[0,0].legend(labels=legend_labels,title='target')

x = iris.pivot(columns='target', values=iris_columns[1])
axs[0,1].hist(x, bins=30)
axs[0,1].set_title("sepal width (cm)")
axs[0,1].set_ylabel("sepal width (cm)")
axs[0,1].legend(labels=legend_labels,title='target')

x = iris.pivot(columns='target', values=iris_columns[2])
axs[1,0].hist(x, bins=30)
axs[1,0].set_title("petal length (cm)")
axs[1,0].set_ylabel("petal length (cm)")
axs[1,0].legend(labels=legend_labels,title='target')

x = iris.pivot(columns='target', values=iris_columns[3])
axs[1,1].hist(x, bins=30)
axs[1,1].set_title("petal width (cm)")
axs[1,1].set_ylabel("petal width (cm)")
axs[1,1].legend(labels=legend_labels,title='target')
plt.show()
```
