### Elbow method

```python
n = 20
k = np.arange(1, n+1, 1)
inertions = []

for i in range(1, n+1):
    kmeans = KMeans(n_clusters=i, init="k-means++").fit(X)
    predicts = kmeans.predict(X)
    inertions.append(kmeans.inertia_)

plt.plot(k, inertions)
plt.xlabel("k")
plt.ylabel("inertia")
plt.title("Inertia based on value of k")
plt.show()
```

### Agglomerative clustering


```python
shopping = pd.read_csv("shopping_data.csv")
shopping.head()
```

```python
shopping.columns
```

```python
data = shopping[['Annual Income (k$)', 'Spending Score (1-100)']]
data.head(5)
```

```python
Z = linkage(data)
dn = dendrogram(Z)
plt.show()
```

```python
methods = ['single', 'average', 'complete', 'ward']
for method in methods:
    Z = linkage(data, method=method)
    dn = dendrogram(Z)
    plt.title(f"dendrogram with {method} linkage")
    plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
for method in methods:
    AggCluster = AgglomerativeClustering(n_clusters=5, linkage=method)
    AggCluster.fit(data)
    predicts = AggCluster.fit_predict(data)
    plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], label=predicts, c=predicts)
    plt.xlabel(f"{data.columns[0]}")
    plt.ylabel(f"{data.columns[1]}")
    plt.title(f"{method} linkage")
    plt.show()
```

### Selecting the number of clusters with silhouette analysis on KMeans clustering


```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
import numpy as np
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)
```

```python
range_n_clusters = [i for i in range(2,7)]
```

```python
plt.scatter(X[:,0], X[:,1], label=y, c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Clusters")
plt.show()
```

```python
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, init='random')
    kmeans.fit(X)
    y_pred = kmeans.predict(X)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, y_pred)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )
    sil_samples = silhouette_samples(X, y_pred)
    y_lower = 10
    fig, (ax1, ax2) = plt.subplots(1,2)
    fig.set_size_inches(16, 6)

    for i in range(n_clusters):
        cluster_values = sil_samples[y_pred==i]
        cluster_values.sort()
        size_cluster_i = cluster_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            cluster_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10
        ax1.set_title(f"The silhouette plot for the {n_clusters} clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        ax1.set_yticks([])
    ax1.axvline(silhouette_avg, linestyle= '--', color='red')

    colors = cm.nipy_spectral(y_pred.astype(float) / n_clusters)
    ax2.scatter(
        X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = kmeans.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

### DBSCAN - density based clusterring


```python
from sklearn.datasets import make_blobs, make_moons, make_circles
x, y = make_moons(n_samples=200, noise=.05, random_state=0)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.3)
dbscan.fit(x)
predicts_dbscan = dbscan.fit_predict(x)

kmeans = KMeans(n_clusters=2, init='random')
kmeans.fit(x)
predicts_kmeans = kmeans.predict(x)

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(16, 6)
ax1.scatter(x[:,0], x[:,1], c=predicts_dbscan)
ax1.set_title("DBSCAN")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")
ax2.scatter(x[:,0], x[:,1], c=predicts_kmeans)
ax2.set_title("kmeans")
ax2.set_xlabel("Feature 1")
ax2.set_ylabel("Feature 2")
plt.show()

```

```python
x, y = make_circles(n_samples=(200,200), noise=.05, random_state=0, factor=0.2)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```
